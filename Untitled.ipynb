```python
import datetime as dt
from matplotlib import pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web

import rule_one_investing as ro
```

```python
style.use('ggplot')
start = dt.datetime(2018,1,1)
end = dt.datetime(2019,6,1)


df = web.DataReader('MSFT','iex', start, end)
plt.plot(df.close)
```

```python
df.tail()
```

```python
closing_prices = list(df.close)
moving_avg = ro.calc_moving_average(closing_prices, steps=5)
exp_moving_avg = ro.calc_exp_moving_average(closing_prices, steps=5)
```

```python
plt.figure(figsize=(30,20))
plt.plot(closing_prices)
plt.plot(moving_avg)
plt.plot(exp_moving_avg)

axes = plt.gca()
max_days = len(closing_prices)
min_days = max_days - 14
# axes.set_xlim([min_days, max_days])

axes.set_ylim([min(closing_prices),max(closing_prices)])
```

```python
closing_prices = list(df.close)

macd, signal_line = ro.calc_macd(closing_prices)
plt.figure(figsize=(30,20))
plt.plot(macd)
plt.plot(signal_line)

axes = plt.gca()
axes.set_ylim([-3, 3])
axes.set_xlim([200, 366])

```

```python
stochastic, stoch_signal_line = ro.calc_stochastic(closing_prices)
plt.figure(figsize=(30,5))
plt.plot(stochastic)
plt.plot(stoch_signal_line)

axes = plt.gca()
axes.set_ylim([-10, 110])
axes.set_xlim([200, 366])
```
